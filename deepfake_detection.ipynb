```python
#frame extraction and storing the frames in frame_output in a new folder made for every mp4 video

import cv2
import os

def extract_frames(video_path, output_folder):
    # Get the base name of the video file without the extension (e.g., id0_0000 from id0_0000.mp4)
    video_name = os.path.splitext(os.path.basename(video_path))[0]
    
    # Create a subfolder for each video inside the main output folder
    video_output_folder = os.path.join(output_folder, video_name)
    os.makedirs(video_output_folder, exist_ok=True)
    
    # Open the video file
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"Error: Could not open video file {video_path}")
        return
    
    count = 0
    while True:
        ret, frame = cap.read()
        
        # If no frame is returned (end of video or error), break the loop
        if not ret:
            print(f"Finished extracting frames for {video_name} or error encountered.")
            break
        
        # Ensure the frame is not empty
        if frame is None:
            print(f"Warning: Empty frame at {count} for {video_name}, skipping.")
            continue
        
        # Save the frame as a .jpg file in the corresponding video subfolder
        frame_filename = os.path.join(video_output_folder, f"frame_{count:05d}.jpg")
        cv2.imwrite(frame_filename, frame)
        count += 1
    
    cap.release()
    print(f"Extracted {count} frames from {video_path} into {video_output_folder}")

# Example usage for multiple videos

#create a method 



video_paths = [
    r'C:\Users\Harshit Agarwal\Desktop\deepfake_detection\Celeb_dataset/fake/id0_id1_0000.mp4',
    r'C:\Users\Harshit Agarwal\Desktop\deepfake_detection\Celeb_dataset/fake/id2_id1_0000.mp4',
    r'C:\Users\Harshit Agarwal\Desktop\deepfake_detection\Celeb_dataset/fake/id3_id2_0000.mp4'
]

for i in range()

output_folder = 'frames_output'

for video_path in video_paths:
    extract_frames(video_path, output_folder)
```

```python
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.models import Model

# Load the EfficientNetB0 model without the top layer (include_top=False)

from tensorflow.keras.applications import Xception

# Load the Xception model with pre-installed weights
base_model = Xception(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Freeze the base model
base_model.trainable = False

# Add custom layers on top
x = GlobalAveragePooling2D()(base_model.output)
x = Dropout(0.5)(x)
x = Dense(1, activation='sigmoid')(x)  # Binary classification: real or fake

# Create the final model
model = Model(inputs=base_model.input, outputs=x)

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Function to preprocess a single frame (resize, normalize)
def preprocess_frame(frame):
    # Resize the frame to 224x224 (EfficientNet input size)
    resized_frame = cv2.resize(frame, (224, 224))
    # Normalize pixel values between 0 and 1
    normalized_frame = resized_frame / 255.0
    return normalized_frame

# Function to load frames from a directory
def load_frames_from_directory(directory_path):
    frames = []
    # List all files in the directory and sort them to maintain order
    frame_files = sorted([f for f in os.listdir(directory_path) if f.endswith(('.png', '.jpg', '.jpeg'))])
    
    for file_name in frame_files:
        # Full path to the image file
        file_path = os.path.join(directory_path, file_name)
        
        # Read the image
        frame = cv2.imread(file_path)
        if frame is not None:
            # Preprocess the frame and append to the frames list
            processed_frame = preprocess_frame(frame)
            frames.append(processed_frame)
        else:
            print(f"Failed to load {file_name}")
    
    return np.array(frames)  # Convert list of frames to a NumPy array

# Example usage
directory_path = r'C:\Users\Harshit Agarwal\Desktop/deepfake_detection/frames_output/id3_id2_0000'  # Path to your frames directory
processed_frames = load_frames_from_directory(directory_path)

print(f"Total frames loaded: {processed_frames.shape[0]}")
print(f"Shape of a single frame: {processed_frames[0].shape}")
```

```python
# Model prediction (batch-wise prediction)
predictions = model.predict(processed_frames)

# Post-processing results
for idx, pred in enumerate(predictions):
    print(f"Frame {idx}: {'Deepfake' if pred > 0.5 else 'Real'}")
```

```python
import numpy as np

# Model prediction (batch-wise prediction)
predictions = model.predict(processed_frames)

# Smooth predictions using a simple moving average
def smooth_predictions(predictions, window_size=5):
    smoothed = []
    for i in range(len(predictions)):
        start_idx = max(0, i - window_size // 2)
        end_idx = min(len(predictions), i + window_size // 2 + 1)
        avg = np.mean(predictions[start_idx:end_idx])
        smoothed.append(avg)
    return np.array(smoothed)

# Apply temporal smoothing
smoothed_predictions = smooth_predictions(predictions)

# Optional: Apply weights to the smoothed predictions (weights can be customized)
weights = np.linspace(0.5, 1.0, len(smoothed_predictions))  # Example weights
weighted_predictions = smoothed_predictions * weights

# Final decision based on the mean of weighted predictions
final_result = 'Deepfake' if np.mean(weighted_predictions) > 0.5 else 'Real'

print(f"Final result for the video: {final_result}")
```
